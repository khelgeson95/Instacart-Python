### Step 2: Find the aggregated mean of the "order_number" column grouped by "department_id" for entire dataframe.

```python
# group ords_prods_merge dataframe by "product_name"
ords_prods_merge.groupby('product_name')
```

```python
# group ords_prods_merge dataframe by "department_id" and calculate the mean for each "order_number" column
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

Since the mean is so high for the missing category this suggests there are a large amount of unspecified values for the column "department_id". The next largest mean value is for the "bulk" department.

### Step 4: Create a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Group data by "user_id" column, generate maximum orders for each user and create "max_order" column for aggregation results
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# tell pandas not to assign a maximum number of rows to display
pd.options.display.max_rows = None
```

```python
# check accuracy of data
ords_prods_merge.head(100)
```

```python
# create loyalty flags to assign labels to user_id based on max_order value
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# check frequency of loyalty flags for accuracy
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

### Step 5: Find the mean of product prices by loyalty group.

```python
# Group ords_prods_merge dataframe by "prices"
ords_prods_merge.groupby('prices')
```

```python
# Group data by "loyalty_flag" and calculate the mean for each "prices" column
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

### Step 6: Create spending flag for each user_id based on average price of products across each user_id's orders.

```python
# Group data by "user_id" column, generate mean product price for each user, create "mean_price" for aggregation results
ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# check output of ords_prods_merge with new column
ords_prods_merge.head(100)
```

```python
# Create flags to assign spending labels to user_id based on mean product price
ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check frequencies of spending flags for accuracy
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

### Step 7: Create an order frequency flag that marks the regularity of a user's ordering behavior according to the median in the "days_since_prior_order" column.

```python
# Group data by "user_id" column, generate median product of "days since prior order, create "median_days" for aggregation results
ords_prods_merge['median_days'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# check output of ords_prods_merge after new "median_days" column
ords_prods_merge.head(100)
```

```python
# Create flags to assign freqency labels to user_id based on median days since prior order
ords_prods_merge.loc[ords_prods_merge['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days'] > 10) & (ords_prods_merge['median_days'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check frequencies of frequency flags for accuracy
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```

```python
# Check the null values
print(ords_prods_merge[ords_prods_merge.frequency_flag.isnull()])
```

#### The null values are due to the customer 159838 only having one order so the median_days cannot be calculated. Therefore this customer could be considered a "Non-frequent customer".

```python
# Change the null values in "frequency_flag" column to "Non-frequent customer"
ords_prods_merge['frequency_flag'].fillna("Non-frequent customer", inplace = True)
```

```python
# check frequency of frequency flags after correcting the null values
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
