# 03. Exercise Example Work

```python
# create a subset of df_ords_prods_merged that includes the first one million rows
df = df_ords_prods_merged[:1000000]
```

```python
# define function for a 'price_label' column
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# apply function to df subset
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# check values in new 'price_range' column
df['price_range'].value_counts(dropna = False)
```

```python
# check the maximum price within the df subset
df['prices'].max()
```

### Do the same process with loc() function instead

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Step 2. Create for-loop to make a new column "busiest days"
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value in [0, 1]:
    result.append("Busiest days")
  elif value in [4, 3]:
    result.append("Least busy days")
  else:
    result.append("Regularly busy")
```

```python
# print result from previous for-loop
result
```

```python
# create a new column within df_ords_prods_merged dataframe and set equal to 'result'
df_ords_prods_merged['busiest_days'] = result
```

```python
# Step 3. Check frequency of 'busiest days' column for accuracy
df_ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

### Step 3. The busiest days count of 11864412 is accurate according to the frequency of 'order_day_of_week' column performed previously. The least busy days count of 7624336 is also accurate. This makes it clear that the least busy days are significantly less busy than either the busiest days or the regularly busy days.

```python
# print the frequency of 'order_hour_of_day' column. Note 0 = 12AM.
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

### Client wants to identify busiest hours of the day with labels "Most orders", "Average orders", "Fewest orders". "Most orders" = [10, 11, 14, 15, 13, 12, 16, 9]. "Average orders" = [17, 8, 18, 19, 20, 7, 21, 22]. "Fewest orders" = [23, 6, 0, 1, 5, 2, 4, 3]

```python
# Step 4. Create for-loop to make a new column "busiest_period_of_day"
result_2 = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_2.append("Most orders")
  elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
    result_2.append("Average orders")
  else:
    result_2.append("Fewest orders")
```

```python
# print result_2 from previous for-loop
result_2
```

```python
# Step 4. Create a new column within df_ords_prods_merged dataframe and set equal to 'result_2'
df_ords_prods_merged['busiest_period_of_day'] = result_2
```

```python
# Step 5. Print frequency of 'busiest_period_of_day' column for accuracy
df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```
